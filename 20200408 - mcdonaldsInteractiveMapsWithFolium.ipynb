## Importing Libraries

```
import pandas as pd
import folium
from folium import plugins
from folium.plugins import MarkerCluster
```

## Importing the Dataset

```

file = '/content/drive/My Drive/MY ANALYSIS/20200408 - McDonalds Localization/Dataset/McDonalds.csv'
columns = ['coordinates','address','building','city','driveHoursFriday','driveHoursMonday','driveHoursSaturday','driveHoursSunday','driveHoursThursday','driveHoursTuesday','driveHoursWednesday','id','identifierValue','gblnumber','storeIdentifier','openstatus','postcode','restaurantUrl','hoursFriday','hoursMonday','hoursSaturday','hoursSunday','hoursThursday','hoursTuesday','hoursWednesday','subDivision','telephone']
dfMcdonalds = pd.read_csv(file, header=0, names=columns)
dfMcdonalds.head()
```

## Creating two new cloumns called Latitude and Longitude

```
dfCoordinates = []
dfCoordinates = dfMcdonalds.coordinates.str.replace(']',"").str.replace('[',"").str.replace(' ',"").str.split(',')
```

```
arrayLatitude = []
arrayLongitude = []
i=0
for i in range(len(dfCoordinates)):
  coordinates = dfCoordinates[i]
  arrayLongitude.append(float(coordinates[0]))
  arrayLatitude.append(float(coordinates[1]))
  i+=1
dfMcdonalds['latitude']=arrayLatitude
dfMcdonalds['longitude']=arrayLongitude
dfMcdonalds.head()
```

## Using Lybrary Folium

### Heat Map

Our dataset there a huge informations about stores, because of this if you put the hole dataset your visualization will be overcrowded. Thus, we'll take a sample of 20% from our Original Dataset.

```
dfMcdonalds1 = dfMcdonalds.sample(frac=0.2)
```

```
coord = []
for lat,lng in zip(dfMcdonalds1.latitude.values[:], dfMcdonalds1.longitude.values[:]):
  coord.append([lat,lng])
```

```
euaHeatMap = folium.Map(location=[55.0000000, -97.0000000], zoom_start=4, tiles='OpenStreetMap')
```

```
euaHeatMap.add_child(plugins.HeatMap(coord))
euaHeatMap
```

### Map with Marker Cluster

```
euaMarker = folium.Map(location=[55.0000000, -97.0000000], zoom_start=5, tiles='cartodbdark_matter')

mc = MarkerCluster()

for i, c in dfMcdonalds.iterrows():
  mc.add_child(folium.Marker([c['latitude'], c['longitude']], 
              popup=str(c['address']),
              tooltip=c['building'],
              icon=folium.Icon(icon='book'))).add_to(euaMarker)
euaMarker
```
